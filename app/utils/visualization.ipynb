```python
import sqlite3
import json
import folium

con = sqlite3.connect('/root/AiSK-backend/restaurant_info.db')

# get all the restaurants
cur = con.cursor()
cur.execute('SELECT * FROM restaurants')
rows = cur.fetchall()
print(len(rows))
# cur.execute('SELECT * FROM MenuItems')
# rows = cur.fetchall()
# print(len(rows))

info_rest=[]
for row in rows:
    info_rest.append({'name': row[1], 'latitude': row[3], 'longitude': row[4]})

# plot all the restaurants location in a map
m = folium.Map(location=[info_rest[0]['latitude'],info_rest[0]['longitude']], zoom_start=12)
  
for i in range(len(info_rest[:500])):
    folium.Marker([info_rest[i]['latitude'],info_rest[i]['longitude']], popup=info_rest[i]['name']).add_to(m)
  
con.close()
    
m
```
